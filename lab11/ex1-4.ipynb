```python
from pymongo import MongoClient
import datetime
```

```python
client = MongoClient("mongodb://localhost")
collection = client['newdb']['restaurants']
```

# Ex 1.1

```python
def find_by_cuisines(cuisines):
  for cuisine in cuisines:
    cursor = collection.find({"cuisine": cuisine})
    for i in cursor:
      print(i['cuisine'], i['name'])
```

```python
find_by_cuisines(['Indian'])
```

# Ex 1.2

```python
find_by_cuisines(["Indian", "Thai"])
```

# Ex 1.3

```python
def restaurant_by_address(building, street, zipcode):
  cursor = collection.find({ 'address.building': building, 'address.street': street, 'address.zipcode': zipcode })
  for i in cursor:
    print(i['name'], i['address'])
```

```python
restaurant_by_address('1115', 'Rogers Avenue', '11226')
```

```python
def delete_single_restuarant_by_borough(borough):
    collection.delete_one({'borough': borough})
```

```python
def delete_all_restuarants_by_cuisine(cuisine):
    collection.delete_many({'cuisine': cuisine})
```

# Ex 4

```python
def update_restaurants():
  result = collection.find({'address.street': 'Rogers Avenue'})
  for restaurant in result:
    C_number = 0
    for grade_data in restaurant['grades']:
      if grade_data['grade'] == 'C':
        C_number += 1

    if C_number > 1:
      res = collection.delete_many({'restaurant_id': restaurant['restaurant_id']})
      print(res.deleted_count, 'deleted', restaurant['name'])
    else:
      new_grades = restaurant['grades']
      new_grades.append({'grade': 'C', 'score': 0, 'date': datetime.datetime.now()})
      res = collection.update_one({"_id": restaurant['_id']}, {'$set': {'grades': new_grades}})
      print(res.modified_count, 'modified', restaurant['name'])
```
